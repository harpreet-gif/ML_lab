#### WAP to perform all dictionary operations

```python
## Creating a dictionary
student = {
    "Name" : "Harpreet Kaur",
    "Rollno" : 3461,
    "Subject" : "Machine Learning",
    "Marks" : 78
}
```

```python
## Display dictionary
print(student)
```

```python
## Access values using keys
print("Name:", student["Name"])
print("Rollno:", student["Rollno"])
print("Subject:", student["Subject"])
```

```python
## Update values
student["Marks"] = 85
print("Updates Marks:", student["Marks"])
```

```python
## Add new key value pair
student["Grade"] = "A"
print(student)
```

```python
## display all keys
print("\nKeys:", student.keys())
```

```python
## display all values
print("Values:", student.values())
```

```python
## Display key-value pairs
print("Items:", student.items())
```

```python
## Length of dictionary
print("Total keys:", len(student))
```

#### Put data into dictionary

```python
names = ["Riya", "Kiran", "Siya", "Palak", "Ajay"]
rollno = [1, 2, 3, 4, 5]
subjects = ("Maths", "Science", "English", "C++", "Java")
marks = [54, 65, 85, 79, 89]

data = {}

for i in range(len(names)):
    data[rollno[i]] = {
        "name": names[i],
        "subject": subjects[i],
        "marks": marks[i]
    }
for rollno, info in data.items():
    print("Roll No:", rollno)
    print("Name:", info["name"])
    print("Subject:", info["subject"])
    print("Marks:", info["marks"])
    print("-" * 25)
```
