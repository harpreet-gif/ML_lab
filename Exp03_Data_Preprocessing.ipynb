## Data Preprocessing

```python
import numpy as np
import pandas as pd
```

```python
raw_data = {
    "Id" : [1, 2, 3, 4, 5, 5],
    "Name" : ["Preet", "Babita", "Pari", "Riya", "Aman", "Aman"],
    "Age" : [19, np.nan, 10, 35, 28, 28],
    "Gender" : ["F", "F", "F", np.nan, "M", "M"],
    "Salary" : [50000, 60000, np.nan, 8000, 55000, 55000],
    "City" : ["New York", "London", "London", "Paris", np.nan, np.nan]
}
df = pd.DataFrame(raw_data)
df
```

## Removing duplicates

```python
df.drop_duplicates()
```

## Handling missing values

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df
```

```python
df["City"] = df["City"].fillna(df["City"].mode()[0])
df
```

```python
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
df
```

## Data type conversion

```python
df["Age"] = df["Age"].astype(int)
df
```

```python
df["Salary"] = df["Salary"].astype(int)
df
```

## Encode Categorical variables

```python
df["Gender"] = df["Gender"].map({"M":0, "F":1})
df
```

```python
df = pd.get_dummies(df, columns = ["City"])
df
```

## Feature Scaling

```python
df["Age.scaled"] = (df["Age"] - df["Age"].mean()) / df["Age"].std()
df
```

```python
df["Salary.scaled"] = (df["Salary"] - df["Salary"].mean()) / df["Salary"].std()
df
```

```python
# Normalization X' = X - Xmin / Xmax-Xmin

df["Salary_norm"] = (df["Salary"] - df["Salary"].min()) / (df["Salary"].max() - df["Salary"].min())
df
```

```python
df["Age_norm"] = (df["Age"] - df["Age"].min()) / (df["Age"].max() - df["Age"].min())
df
```

## Outlier detection & Removal (IQR method)

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)
IQR = Q3 - Q1
df = df[(df["Salary"] >= Q1 - 1.5 * IQR) & (df["Salary"] <= Q3 + 1.5 * IQR)]
df
```

## Feature Selection (drop unnecessary columns)

```python
df = df.drop(["Name"], axis = 1)
df
```

```python
df.reset_index(drop = True, inplace = True)
df
```
