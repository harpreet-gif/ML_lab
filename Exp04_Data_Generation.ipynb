```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(17)   ## for reproducibility
n = 20
num = np.arange(1, n+1)
num[-2] = num[-3]    ## duplicates (replace 19 with 18)
print(num)
print(num[-2])
print(num[-8])
```

#### 1. For generating random names for given list.

```python
names_list = ["Harpreet", "Riya", "Ayra", "Zuraiz", "Rahul", "Babita", "Romaila"]
names = np.random.choice(names_list, n)
print(names)
```

#### 2. For generating random age 

```python
age_list = (19, 21, 27, 20, 28, 26, 31, 30)
age = np.random.choice(age_list, n)
print(age)
## age_list = np.random.randint(19, 35, size = n)
## age_list
```

#### 3. For generating random gender using p variable

```python
gender = np.random.choice(["M", "F", np.nan], size = n, p = [0.50, 0.45, 0.05])
gender
```

#### 4. Generating age with randint and as type function

```python
age_list = np.random.randint(19, 35, size = n).astype(float)
age_list[[1,6]] = np.nan
age = age_list
age
```

#### 5. Generating salary with randint and astype function. 

```python
salary_list = np.random.randint(21000, 55000, size = n).astype(float)
age_list[[2,8]] = np.nan
salary = salary_list
salary
```

#### 6. Generating dataframe

```python
id = list(range(1, len(names)+ 1))
data = pd.DataFrame({"Id" : id, "Name": names, "Age": age, "Gender": gender, "Salary": salary})
data
```
